```python
# Required Libraries for data extraction and insertion into SQL database

import os

# SQL and Pandas libraries

import mysql.connector 
import json
import pandas as pd
```

```python
# Connecting to MySQL database

conn = mysql.connector.connect(host='127.0.0.1',
                               user='root',
                               password='root',
                               database= 'Phonepe_Pulse',
                               auth_plugin='mysql_native_password'
                              )
                                                                                                                                                     
cursor = conn.cursor()
```

```python
#Aggregate Transaction

path="D:/data science - guvi/MDT-34/capstone project/Project 2 - Phonepe Pulse Data Visualization and Exploration/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)


clm={'State':[], 'Year':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transaction_type'].append(Name)
              clm['Transaction_count'].append(count)
              clm['Transaction_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Trans=pd.DataFrame(clm)
```

```python
#Inserting into Aggregate Transaction table

insert_query = """INSERT INTO Aggregate_Transaction (`State`, `Year`, `Quarter`, `Transaction_type`, `Transaction_count`,`Transaction_amount`)
              VALUES (%s, %s, %s, %s, %s,%s)"""

for i in Agg_Trans .to_dict("records"):
     cursor.execute(insert_query,(i['State'],i['Year'],i['Quarter'],i['Transaction_type'],i['Transaction_count'],i['Transaction_amount']))
conn.commit()
```

```python
# Aggregate User 

path = "D:/data science - guvi/MDT-34/capstone project/Project 2 - Phonepe Pulse Data Visualization and Exploration/pulse/data/aggregated/user/country/india/state/"

Agg_state_list = os.listdir(path)

clm1 = {'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Count_of_users_by_this_brand': [],'Percentage': []}

clm2 = {'State': [], 'Year': [], 'Quarter': [], 'Registered_Users':[], 'No_of_App_Opens':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)

        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            
            if D["data"]["usersByDevice"]:
                for z in D["data"]["usersByDevice"]:
                    brand_name = z.get("brand","Not Available")
                    counts = z.get("count",0)
                    percents = z.get("percentage",0)
                    clm1["Brands"].append(brand_name)
                    clm1['Count_of_users_by_this_brand'].append(counts)
                    clm1["Percentage"].append(percents)
                    clm1["State"].append(i)
                    clm1["Year"].append(j)
                    clm1["Quarter"].append(int(k.strip('.json')))
                    
            for y in D["data"]:
                    if "aggregated" in D["data"]:
                        aggregated_data = D["data"]["aggregated"]
                        registeredUsers = aggregated_data.get("registeredUsers", 0)
                        appOpens = aggregated_data.get("appOpens", 0)
                    
                        clm2['Registered_Users'].append(registeredUsers)
                        clm2['No_of_App_Opens'].append(appOpens)
                    
                        clm2["State"].append(i)
                        clm2["Year"].append(j)
                        clm2["Quarter"].append(int(k.strip('.json')))
        
           
#Succesfully created a dataframe

Agg_user_brand= pd.DataFrame(clm1)
Agg_user_quarter = pd.DataFrame(clm2)
```

```python
# Top Transaction (District and pincode)

path="D:/data science - guvi/MDT-34/capstone project/Project 2 - Phonepe Pulse Data Visualization and Exploration/pulse/data/top/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list

clm2={'State':[], 'Year':[],'Quarter':[] , 'District':[] , 'District_Transaction_count':[], 'District_Transaction_amount':[] }
clm3={'State':[], 'Year':[],'Quarter':[] ,'Pincode':[] , 'Pincode_Transaction_count':[], 'Pincode_Transaction_amount':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['pincodes']:
              
              pincodes_entityName = z['entityName']
              pincodes_count = z['metric']['count']
              pincodes_amount = z['metric']['amount']

              clm3['Pincode'].append(pincodes_entityName)
              clm3['Pincode_Transaction_count'].append(pincodes_count)
              clm3['Pincode_Transaction_amount'].append(pincodes_amount)
              clm3['State'].append(i)
              clm3['Year'].append(j)
              clm3['Quarter'].append(int(k.strip('.json')))
            
            
            for x in D['data']['districts']:

              districts_entityName = x['entityName']
              districts_count = x['metric']['count']
              districts_amount = x['metric']['amount']

              clm2['District'].append(districts_entityName)
              clm2['District_Transaction_count'].append(districts_count)
              clm2['District_Transaction_amount'].append(districts_amount)
              clm2['State'].append(i)
              clm2['Year'].append(j)
              clm2['Quarter'].append(int(k.strip('.json')))

#Succesfully created a dataframe 
Top_Trans_pincode=pd.DataFrame(clm3)
Top_Trans_District=pd.DataFrame(clm2)
```
